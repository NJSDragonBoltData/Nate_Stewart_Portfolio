```python
# Question: Build out Career_length Column
# Hint: Convert rookie/final year and subtract

df['Career_length (in years)'] = df['Final_Year'] - df['Rookie_Year']

df
```

```python
# Calculations with the dataframe

# Question 1: Cricketers in this dataframe. What is the average career length?

df['Career_length (in years)'].mean()
```

```python
# Question 2: Average Batting_Strike_Rate for Cricketers who played over the past 10 years.

df[df['Career_length (in years)'] > 10]['Batting_Strike_Rate'].mean()
```

```python
# Question 3: Find number of cricketers who played before 1960.

df[df['Rookie_Year'] < 1960]['Player'].count()
```

```python
# Question 4: Maximum Highest_Inns_Score by country.

df.groupby('Country')['Highest_Inns_Score'].max()
```

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('High_Inn_Country').reset_index().sort_values('High_Inn_Country', ascending = False)
```

```python
# Question 5: Hundreds, Fifties, ducks (0) AVG by country.

df.groupby('Country')[['100', '50', '0']].mean()
```
