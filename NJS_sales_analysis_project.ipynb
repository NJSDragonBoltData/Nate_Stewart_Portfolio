```python
# Importing libraties

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import CSV or Excel file for data analysis

df = pd.read_excel('superstore_sales.xlsx')

df
```

```python
# Summarizing the entire dataset

df.info()
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Dropping null values just for show

new_df = df.dropna()

new_df
```

```python
# Descriptive statistics without order_date and ship_date

df_dropdates = new_df.drop(['order_date', 'ship_date'], axis=1)

df_dropdates.describe()
```

```python
#### Which country has generated the most in sales?

pd.DataFrame(df['country'].value_counts())
```

```python
country_sales = pd.DataFrame(df.groupby('country')['sales'].nunique())

country_sales_sort = country_sales.sort_values('sales',ascending=False)

country_sales_sort
```

```python
# Top 10 countries that generated the most sales

top_10_sales = country_sales_sort[:10]

top_10_sales
```

```python
### What is the overall sales trend?
```

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Finding the month year from the dataset

df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))

df['month_year']
```

```python
# Grouping the month year
sales_trend = df.groupby('month_year').sum(numeric_only=True)['sales'].reset_index()
```

```python
# Visualizing sales trend

plt.figure(figsize=(20, 6))
plt.plot(sales_trend['month_year'], sales_trend['sales'])
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
product_sales = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['sales'])
top_product_sales_sort = product_sales.sort_values('sales', ascending=False)
top_product_sales_sort[:10]
```

```python
### What are the most selling products

most_selling_prod = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['quantity'])
```

```python
top_most_selling_prod = most_selling_prod.sort_values('quantity', ascending=False)

top_most_selling_prod[:10]
```

```python
### What is the most preferred shipping method

sns.countplot(x=df['ship_mode'], palette='plasma')
plt.figure(figsize=(12,10))
plt.show()
```

```python
### Find the most profitable category and subcategory

profit_cat_subcat = pd.DataFrame(df.groupby(['category','sub_category']).sum(numeric_only=True)['profit'])

profit_cat_subcat.sort_values(['category','profit'], ascending=False)
```
